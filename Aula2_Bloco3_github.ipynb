Estruturas de iteração, repetição, laço ou loop.

### while e for

As estruturas de repetição, também conhecidas como **loops ou laços**, são instruções que permitem executar um bloco de código várias vezes seguidas até que uma determinada condição seja atendida ou que uma sequência chegue ao fim.

## Estrutura While

O **while** executa um bloco de código repetidamente enquanto uma condição específica for verdadeira

```python
print("Olá!")
print("Olá!")
print("Olá!")
print("Olá!")
```

```python
cont = 1 # variável contador para controlar quantas vezes o loop será executado (start)
while cont <= 10:
  print("Olá!")
  cont += 1        # operador de atribuição aumentada: cont = cont + 1
print("FIM")
```

enquanto uma condição
(ou um conjunto delas) for verdadeira, uma instrução (ou
um conjunto de instruções) que está dentro do laço deverá ser

```python
cont = 1 # contar a quantidade de vezes que o laço while está sendo executado
while cont <= 10: # quando o valor de cont passar a ser maior que 10, a condição de execução deixará de ser verdadeira e, consequentemente, as instruções definidas nas linhas 3 e 4 não serão executadas
  print("Olá!")
  cont += 1 #todas as vezes que o programa entrar no laço, a variável contadora terá o seu valor aumentado em 1
print("FIM")
```

## Função Range

É uma função utilizada para gerar uma **sequência de números inteiros**.

Ela é amplamente usada em estruturas de repetição (como o loop for) para controlar o número de vezes que um bloco de código deve ser executado ou para iterar sobre índices

A função range() gera uma sequência de números dentro de
uma faixa especificada.

A função range() recebe um
parâmetro o qual indica a quantidade N de números a serem
gerados, onde N deve ser maior que 0.

```python
list(range(5))
```

```python
list(range(10))
```

Range tem três parâmetros distintos:

* início_da_faixa

* fim_da_faixa

* incremento




```python
list(range(10, 16)) #Incremento = 1
```

```python
list(range(10, 16, 2)) #Incremento = 2:
```

```python
list(range(16, 10, -1)) #Incremento = -1:
```

## Estrutura For

**For** é usado quando você sabe exatamente quantas vezes deseja repetir algo ou quando precisa percorrer cada item de uma coleção (como listas, strings ou dicionários)

while ou for?

**While** - iterações (loops) indefinidas (repetição até atingir uma condição).

**For** - iterações (loops) definidas (percorrer uma lista conhecida)

```python
for cont in range(10):
  print("Olá!")
print("FIM")
```

## Comando Break

O objetivo é **quebrar** a execução de uma estrutura de repetição, isto
é, forçar a saída do fluxo do programa de dentro do laço

```python
# Total abastecido e em quantos minuto?

capacidade = 10000  # litros
abastecido = 0
taxa = 800  # litros por minuto
minuto = 0

print("Iniciando abastecimento...\n")

while True:     # Uso do while True: Cria um loop infinito que só termina com break.
    minuto += 1
    abastecido += taxa

    if abastecido >= capacidade:
        print(f"Minuto {minuto}: {abastecido} litros - CAPACIDADE ATINGIDA!")
        print("Parando o abastecimento para evitar transbordamento.")
        break

    print(f"Minuto {minuto}: {abastecido} litros")

print(f"\nTotal abastecido: {abastecido} litros em {minuto} minutos.")
```

## Exercício 5

A linha A124 registrou os seguintes embarques em dias úteis.
Percorra a lista e calcule:

* O total de passageiros na semana

* Quantos dias tiveram mais de 8500 passageiros

```python
embarques_semana = [8200, 7900, 8450, 9100, 8800, 8500]
```

```python
# Exercício 5

total = 0
dias_movimento_alto = 0

for passageiros in embarques_semana:
    total += passageiros             # total = total + passageiros

    if passageiros > 8500:
        dias_movimento_alto += 1

print("Total de passageiros na semana:", total)
print("Dias com mais de 8500 passageiros:", dias_movimento_alto)
```

## Exercício 6

Agora imagine que um ônibus articulado dessa linha tem capacidade média de 120 passageiros por viagem.

Quantas viagens são necessárias para transportar 3000 passageiros em um terminal.

```python
capacidade_onibus = 120
passageiros_terminal = 3000

transportados = 0
viagens = 0

while transportados < passageiros_terminal:
    transportados += capacidade_onibus      # transportados = transportados + capacidade_onibus
    viagens += 1                            # viagens = viagens + 1

print("Viagens necessárias:", viagens)
print("Passageiros transportados:", transportados)
```

## Exercício 9

Imagine um aplicativo de transporte (como Uber, 99 ou táxi) que mantém uma lista de passageiros aguardando viagem. O motorista está procurando por um passageiro específico que pediu uma corrida com prioridade ("VIP"). Assim que encontrar o primeiro passageiro VIP na lista, ele deve parar a busca e atender esse passageiro.

```python
fila_espera = [
    ("Carlos", "comum"),
    ("Ana", "comum"),
    ("João", "VIP"),
    ("Maria", "comum"),
    ("Pedro", "VIP"),
    ("Lucia", "comum")
]

print("Motorista procurando o primeiro passageiro VIP na fila...\n")

posicao = 0  # contador manual (começa em 0, mas vamos exibir posição+1)

for passageiro in fila_espera:
    posicao += 1       # posicao = posicao + 1
    nome, prioridade = passageiro

    if prioridade == "VIP":
        print(f"Passageiro VIP encontrado na posição {posicao}: {nome}")
        print("Motorista irá atender este passageiro imediatamente.")
        break

    print(f"Passageiro {nome} é comum, continuando busca...")

print("\nBusca encerrada.")
```

O **enumerate** é uma função embutida do Python que permite percorrer uma sequência (como uma lista) e, a cada iteração, obter tanto o índice (posição) quanto o valor do elemento atual.

A linha nome, prioridade = passageiro é uma forma elegante e legível de desempacotar (ou desestruturar) uma tupla.

No código, passageiro é uma tupla com dois elementos: o nome e a prioridade. Em vez de acessar cada elemento por índices (como passageiro[0] e passageiro[1]), o Python permite atribuir diretamente esses valores a duas variáveis separadas em uma única linha.

Em Python, a principal diferença prática entre tupla e lista é a mutabilidade:

Listas são mutáveis → você pode adicionar, remover ou alterar elementos depois de criadas.

Tuplas são imutáveis → depois de criadas, não podem ser modificadas.

```python
# Lista de passageiros aguardando
fila_espera = [
    ("Carlos", "comum"),
    ("Ana", "comum"),
    ("João", "VIP"),
    ("Maria", "comum"),
    ("Pedro", "VIP"),
    ("Lucia", "comum")
]

print("Motorista procurando o primeiro passageiro VIP na fila...\n")

for i, passageiro in enumerate(fila_espera):  # Para cada i (index), passageiro na lista sequencial
    nome, prioridade = passageiro # Passageiro é uma tupla e tem dois elementos: nome, prioridade

    if prioridade == "VIP":
        print(f"Passageiro VIP encontrado na posição {i+1}: {nome}")
        print("Motorista irá atender este passageiro imediatamente.")
        break             # Interrompe a busca após encontrar o primeiro VIP

    print(f"Passageiro {nome} é comum, continuando busca...")

print("\nBusca encerrada.")
```

# 8. FUNÇÕES

Uma **função** é um bloco de código que executa uma tarefa específica. Ela pode ser definida uma vez e chamada (usada) várias vezes ao longo do programa. Isso evita repetição, organiza melhor o código e facilita a manutenção

```python
def dizer_ola():
    print("Olá, mundo!")

# Chamando a função
dizer_ola()
```

```python
def saudacao(nome):
    print(f"Olá, {nome}!")

saudacao("Ana")
saudacao("Carlos")
```

```python
def somar(a, b):
    resultado = a + b
    return resultado

# O valor retornado pode ser guardado numa variável
total = somar(5, 3)
print(total)
```

```python
def saudacao(nome, mensagem="Olá"):
    print(f"{mensagem}, {nome}!")

saudacao("João")
saudacao("Maria", "Bom dia")
```

```python
def dobro(num):
    return num * 2

def triplo(num):
    return num * 3

def calcular(operacao, valor):
    if operacao == "dobro":
        return dobro(valor)
    elif operacao == "triplo":
        return triplo(valor)
    else:
        return "Operação desconhecida"

print(calcular("dobro", 5))
print(calcular("triplo", 5))
```

```python
"""Calcula a área de um retângulo."""

def area_retangulo(base, altura):
    return base * altura

print(area_retangulo(2, 4))
```

## Exercício 11

 Crie uma função que receba um valor em reais e uma cotação do dólar, e retorne o valor convertido

```python
def real_para_dolar(valor_reais, cotacao_dolar):
    valor_dolar = valor_reais / cotacao_dolar
    return valor_dolar

reais = float(input("Digite o valor em reais: "))
cotacao = 5.20  # Exemplo: 1 dólar = 5,20 reais

dolares = real_para_dolar(reais, cotacao)

print(f"R$ {reais:.2f} equivalem a US$ {dolares:.2f} (cotação: R$ {cotacao:.2f})")
```
