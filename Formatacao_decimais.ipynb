# Formatação decimais

```python
from babel.numbers import format_currency

lucro = 40000
print(format_currency(lucro, 'BRL', locale='pt_BR'))
# Saída: R$ 40.000,00
```

```python
from babel.numbers import format_decimal

valor = 40000
numero_formatado = format_decimal(valor, locale='pt_BR')
print(f"{numero_formatado}")
# Saída: 40.000
```
