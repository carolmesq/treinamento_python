## Exemplo 1

**Análise de temperatura em diferentes regiões**

Suponha que temos a quantidade de passageiros em três linhas da cidade durante uma semana. Queremos calcular:

* A média passageiros por linha;

* O maior carregamento por linha;

* O dia maior movimento de cada linha.

```python
# Dados simulados: passageiros transportados por linha

passageiros = np.array([
    [1200, 950, 800],   # Segunda
    [1300, 1020, 1250],  # Terça
    [1250, 980, 820],   # Quarta
    [1400, 850, 900],  # Quinta
    [1500, 1150, 950],  # Sexta
    [1100, 880, 750],   # Sábado
    [900,  700, 600]    # Domingo
])

linhas = ['Linha Azul', 'Linha Vermelha', 'Linha Verde']
```

```python
# Cálculos
media_linha = np.mean(passageiros, axis=0)
maior_carregamento = np.max(passageiros, axis=0)
dia_maior_movimento = np.argmax(passageiros, axis=0)

print("Média de passageiros por linha:")
for i, linha in enumerate(linhas):
    print(f"{linha}: {media_linha[i]:.0f} passageiros")

print("\nO maior carregamento por linha:")
for i, linha in enumerate(linhas):
    print(f"{linha}: {maior_carregamento[i]:.0f} passageiros")

print("\nDia de maior movimento (0=segunda, 6=domingo):")
for i, linha in enumerate(linhas):
    print(f"{linha}: {dia_maior_movimento[i]}")
```

## Exemplo 2

**Dados de população de bairros ao longo dos anos**

Temos uma matriz onde as linhas representam bairros e as colunas representam anos (2018, 2019, 2020). Queremos:

* O crescimento populacional de cada bairro (diferença
entre último e primeiro ano);

* O ano com maior população total (soma de todos os bairros).

```python
pop_bairros = np.array([[4500, 5200, 4800],   # Bairro Centro
                        [6100, 5900, 6300],   # Bairro Recanto
                        [7200, 7500, 7800]])  # Bairro Mantiqueira
anos = [2018, 2019, 2020]
bairros = ['Centro', 'Recanto', 'Mantiqueira']

crescimento = pop_bairros[:, -1] - pop_bairros[:, 0]
print("Crescimento populacional 2018-2020 por bairro:")
for i, cresc in enumerate(crescimento):
    print(f"{bairros[i]}: {cresc} habitantes")

# População total por ano
pop_anual = np.sum(pop_bairros, axis=0)
ano_maior_pop = anos[np.argmax(pop_anual)]  # índice da linha com maior valor
print(f"\nAno com maior população total: {ano_maior_pop}")
```

## Exemplo 3

**Simulação de dados de tráfego**

Vamos gerar dados simulados de contagem de veículos em 5 cruzamentos, a cada hora, durante um dia (24 horas). Depois, calcular:

* O cruzamento mais movimentado (média horária mais alta).

* O horário de pico (hora com maior fluxo médio entre cruzamentos).

```python
# Gerando dados aleatórios (entre 10 e 300 veículos por hora)
np.random.seed(42)  # define um valor inicial para o gerador de números aleatórios do NumPy
fluxo = np.random.randint(10, 300, size=(5, 24))  # 5 cruzamentos, 24 horas /// randint = gera arrays preenchidos com números inteiros aleatórios dentro de um intervalo específico

# Média por cruzamento
media_cruzamento = np.mean(fluxo, axis=1)
cruzamento_mais_mov = np.argmax(media_cruzamento)
print(f"Cruzamento mais movimentado: {cruzamento_mais_mov} (média de {media_cruzamento[cruzamento_mais_mov]:.0f} veículos/hora)")

# Média por hora (entre todos os cruzamentos)
media_hora = np.mean(fluxo, axis=0)
hora_pico = np.argmax(media_hora)
print(f"Horário de pico: {hora_pico}h (média de {media_hora[hora_pico]:.0f} veículos)")

# Horas em que o fluxo médio ultrapassa 200 veículos
horas_criticas = np.where(media_hora > 200)[0]
print(f"Horas com média >200: {horas_criticas}")
```

**Estruturas de dados: Series e DataFrame**

* Series: é como uma lista com índice, unidimensional.

* DataFrame: é uma tabela com linhas e colunas, semelhante a uma planilha ou tabela SQL.

```python
# Criando uma Series simples
s = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
print(s)
```

Principais diferenças entre **Series** (Pandas) e **Arrays** (NumPy):
* Índices e Rótulos: **Series** possuem um índice explícito (que podem ser rótulos, datas, etc.) que permite alinhamento automático de dados, enquanto **arrays** NumPy usam apenas posicionamento inteiro (0, 1, 2...).
* Tipos de Dados: **Arrays NumPy** geralmente contêm dados do mesmo tipo (homogêneos) para maior desempenho, enquanto **Series** podem conter tipos mistos (heterogêneos), embora funcionem melhor com tipos consistentes.
* Uso de Memória e Desempenho: **Arrays NumPy** são mais eficientes em memória e mais rápidos para cálculos numéricos puros. O **Pandas** é mais flexível, porém consome mais memória.
* Dimensões: **Series** é estritamente unidimensional. **Arrays NumPy** (ndarray) podem ser unidimensionais, bidimensionais ou multidimensionais (matrizes).

## Exemplo 4

Temos um arquivo registros_onibus.csv com dados de viagens de ônibus (data, linha, embarques, desembarques, tempo_viagem, atraso_min). Queremos responder:

* Qual a média de embarques por linha?

* Em quais dias houve mais passageiros?

* Relação entre tempo/viagem e embarques?

```python
import pandas as pd
import numpy as np

# Gerar dados sintéticos
np.random.seed(42)
datas = pd.date_range('2024-01-01', periods=30, freq='D')
linhas = ['Amarela', 'Azul', 'Rosa']
dados = []

for data in datas:
    for linha in linhas:
        embarques = np.random.randint(200, 500)
        desembarques = embarques - np.random.randint(0, 50)
        tempo_viagem = np.random.randint(18, 45)
        atraso = np.random.randint(0, 15)
        dados.append([data, linha, embarques, desembarques, tempo_viagem, atraso])

df = pd.DataFrame(dados, columns=['data', 'linha', 'embarques', 'desembarques', 'tempo_viagem', 'atraso_min'])
df.to_csv('registros_onibus.csv', index=False)
```

```python
df = pd.read_csv('registros_onibus.csv')
df['data'] = pd.to_datetime(df['data'])
print(df.info())
```

```python
print(df.describe())
```

```python
# Média de embarques por linha
media_linha = df.groupby('linha')['embarques'].mean()
print("Média de embarques por linha:\n", media_linha)
```

```python
# Dia com maior total de embarques (soma todas as linhas)
total_dia = df.groupby('data')['embarques'].sum()
dia_pico = total_dia.idxmax()
print(f"Dia com mais passageiros: {dia_pico.date()} com {total_dia.max()} embarques")
```

```python
# Correlação entre temperatura e embarques
corr = df['tempo_viagem'].corr(df['embarques'])
print(f"Correlação temperatura vs embarques: {corr:.2f}")
```

```python
# Dias com atraso médio acima de 10 minutos
atraso_medio_dia = df.groupby('data')['atraso_min'].mean()
dias_atrasados = atraso_medio_dia[atraso_medio_dia > 10]
print("Dias com atraso médio > 10 min:\n", dias_atrasados)
```

```python
# Salvando a tabela em csv
resumo = df.groupby('linha').agg({
    'embarques': ['sum', 'mean', 'max'],
    'atraso_min': 'mean'
})
resumo.to_csv('resumo_onibus.csv')
```

* Axes (eixos): a área onde os dados são plotados (geralmente um gráfico com eixos x e y).

Gráfico de Dispersão (Scatter) – Relações entre variáveis

```python
import numpy as np

# Dados sintéticos: horário de partida (em horas) e tempo de viagem (minutos)
np.random.seed(42)
horario_partida = np.random.uniform(6, 22, 100)
tempo_viagem = 20 + 10 * np.sin((horario_partida - 8) * np.pi / 12) + np.random.normal(0, 5, 100)

plt.scatter(horario_partida, tempo_viagem, alpha=0.6, c='blue')
plt.title('Tempo de viagem x horário de partida')
plt.xlabel('Horário de partida (h)')
plt.ylabel('Tempo de viagem (min)')
plt.grid(True, alpha=0.3)
plt.show()
```

Gráfico de Barras – Comparações entre categorias

```python
modais = ['Ônibus', 'Metrô', 'Trem', 'Bicicleta']
passageiros_modal = [1200, 850, 430, 310]  # milhares/dia

plt.bar(modais, passageiros_modal, color=['blue', 'red', 'green', 'orange'])
plt.title('Passageiros diários por modal (milhares)')
plt.xlabel('Modal')
plt.ylabel('Passageiros (mil)')
plt.show()
```

Histograma – Distribuição de uma variável

```python
atrasos = np.random.exponential(scale=5, size=200)  # simulação de atrasos

plt.hist(atrasos, bins=20, edgecolor='black', color='skyblue')
plt.title('Distribuição dos atrasos (minutos)')
plt.xlabel('Atraso (min)')
plt.ylabel('Frequência')
plt.show()
```

Gráfico de Pizza – Proporções

```python
modais = ['Ônibus', 'Metrô', 'Bicicleta', 'A pé']
percentuais = [45, 30, 15, 10]
plt.pie(percentuais, labels=modais, autopct='%1.1f%%', startangle=90)
plt.title('Divisão modal (% das viagens)')
plt.axis('equal')   # garante que o círculo seja desenhado como círculo
plt.show()
```

Subplots – Múltiplos Gráficos em uma Figura

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))   # 2 linhas, 2 colunas

# Gráfico 1: passageiros por hora
axes[0, 0].plot(horarios, passageiros, color='blue')
axes[0, 0].set_title('Passageiros por hora')
axes[0, 0].set_xlabel('Hora')
axes[0, 0].set_ylabel('Passageiros')

# Gráfico 2: dispersão tempo x horário
axes[0, 1].scatter(horario_partida, tempo_viagem, color='red', alpha=0.5)
axes[0, 1].set_title('Tempo de viagem x horário')
axes[0, 1].set_xlabel('Horário')
axes[0, 1].set_ylabel('Tempo (min)')

# Gráfico 3: barras modais
axes[1, 0].bar(modais, passageiros_modal, color=['gold', 'turquoise', 'limegreen', 'violet'])
axes[1, 0].set_title('Passageiros por modal')
axes[1, 0].set_ylabel('Milhares')

# Gráfico 4: histograma atrasos
axes[1, 1].hist(atrasos, bins=15, color='orange', edgecolor='darkgoldenrod')
axes[1, 1].set_title('Distribuição de atrasos')
axes[1, 1].set_xlabel('Atraso (min)')

plt.tight_layout()   # ajusta espaçamento
plt.show()
```

**Dicas** para Gráficos Mais Informativos

* Anotações: destaque pontos importantes com plt.annotate().

* Barras de erro: úteis para mostrar incertezas ou variações.

* Cores: use paletas adequadas (ex.: plt.cm.viridis).

* Tamanho da figura: plt.figure(figsize=(10,6)) para gráficos maiores.

* Salvar figura: plt.savefig('grafico.png', dpi=300).
