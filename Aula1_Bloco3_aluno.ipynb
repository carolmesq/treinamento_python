**Diagnóstico Integrado**

Condição	Resultado Final
* Nenhum CRÍTICO	-> SISTEMA ADEQUADO
* 1 CRÍTICO	-> SISTEMA EM ALERTA
* 2 ou mais CRÍTICOS	-> SISTEMA DEFICIENTE

```python
total_onibus_frota = 9000
total_linhas_onibus = 1173
linhas_noite = 20
horas_operacao_dia = 17

# Cálculos
media_onibus = total_onibus_frota / total_linhas_onibus
percentual_noite = (linhas_noite / total_linhas_onibus) * 100

print("\n--- Avaliação do Sistema ---")

criticos = 0

# Indicador 1
if media_onibus >= 8:
    print("Oferta por linha: BOM")
elif media_onibus >= 5:
    print("Oferta por linha: MÉDIO")
else:
    print("Oferta por linha: CRÍTICO")
    criticos += 1

# Indicador 2
if percentual_noite >= 12:
    print("Cobertura noturna: BOM")
elif percentual_noite >= 8:
    print("Cobertura noturna: MÉDIO")
else:
    print("Cobertura noturna: CRÍTICO")
    criticos += 1

# Indicador 3
if horas_operacao_dia == 24:
    print("Continuidade operacional: BOM")
elif horas_operacao_dia >= 18:
    print("Continuidade operacional: MÉDIO")
else:
    print("Continuidade operacional: CRÍTICO")
    criticos += 1

# Diagnóstico final
print("\n--- Diagnóstico Final ---")

if criticos == 0:
    print("SISTEMA ADEQUADO")
elif criticos == 1:
    print("SISTEMA EM ALERTA")
else:
    print("SISTEMA DEFICIENTE")
```

# Exercício 4

Análise de gasto mensal x passe mensal

```python
# Dados
tarifa = 5.30  # valor da tarifa em reais
dias_uteis = 22
passe_mensal = 220.00  # valor hipotético de um passe mensal

# Entrada do usuário
uso = int(input("Quantas vezes você usa o ônibus por dia? "))

# Validação da entrada (se o número for negativo ou zero)
if uso <= 0:
    print("Você digitou um número inválido. Por favor, reinicie o programa e digite um valor positivo.")
else:
    gasto_dia = uso * tarifa
    gasto_mes = gasto_dia * dias_uteis

    print(f"Gasto por dia: R$ {gasto_dia:.2f}")
    print(f"Gasto mensal (22 dias): R$ {gasto_mes:.2f}")

    # Classificação do gasto mensal
    if gasto_mes < 100:
        print("Seu gasto mensal é baixo.")
    elif gasto_mes < 200:
        print("Seu gasto mensal é moderado.")
    else:
        print("Seu gasto mensal é alto.")

    # Comparação com o passe mensal
    if gasto_mes > passe_mensal:
        economia = gasto_mes - passe_mensal
        print(f"Você economizaria R$ {economia:.2f} comprando um passe mensal!")
    elif gasto_mes < passe_mensal:
        diferenca = passe_mensal - gasto_mes
        print(f"O passe mensal sairia R$ {diferenca:.2f} mais caro que pagar por viagem.")
    else:
        print("Seu gasto mensal é exatamente igual ao valor do passe mensal.")
```
