```python
import chardet  # detectar a codificação de uma sequência de bytes

with open('bairros.shp', 'rb') as f:
    raw = f.read(10000)  # lê os primeiros bytes
    encoding = chardet.detect(raw)['encoding']
    print(encoding)      # ex.: 'ISO-8859-1'

# Depois use essa codificação
bairros_vgp = gpd.read_file('bairros.shp', encoding=encoding)
```
